```python
# Part 1: The Ideal Case

#1. free fall no drag
import numpy as np
import matplotlib.pyplot as plt
g = 9.81 # gravitational constant 9.81 m/s^2
def freefall_nodrag(h):
    return np.sqrt(2*h/g)
test_calculation = freefall_nodrag(4000)
print(f"{test_calculation}(s) to reach 4km bottom of shaft no drag")
```

2.
We start with the equation $\frac{d^{2}y}{dt^{2}} = -g -\alpha (\frac{dy}{dt})^{\gamma}$.  We then let $v = \frac{dy}{dt}$.  Through substitution we are left with a system of coupled ODE's $v = \frac{dy}{dt}$ and $\frac{dv}{dt} = -g -\alpha (v)^{\gamma}$

```python
# 3 
from scipy.integrate import solve_ivp
y = 4000 # (m) depth of the mineshaft
alpha = 0 # drag coefficient
gamma = 0 # velocity dependence on drag
def DerivativesnoDrag(t,s):
    y,v = s
    dydt = v
    dvdt = -g
    return [dydt,dvdt]
def bottomofshaft(t,s):   # creating an event to give me the time when it hits the bottom of the shaft
    return s[0]
bottomofshaft.terminal = False
bottomofshaft.direction = -1

v0 = 0   # intial velocity
y0 = 4000  # depth of the mineshaft
s0=[y0,v0]
t_eval = np.linspace(0,30,50)
soln = solve_ivp(fun=DerivativesnoDrag,t_eval=t_eval,t_span=(0,30),y0=s0,events=bottomofshaft)
print(soln)
t_event = soln.t_events[0][0]
print(t_event)
t = soln.t
y1 = soln.y[0]
y2 = soln.y[1]
fig, ax = plt.subplots()
ax.plot(t,y1,color = 'red',label="Position")
ax.set_xlabel("Time (s)")
ax.set_title("Free Fall no Drag (Figure 1)")
ax.grid()
ax2 = ax.twinx()
ax2.plot(t,y2,color = 'purple',label="Velocity")
ax.axhline(0,color = 'black')
ax.set_ylabel("Position (m)",color='red')
ax2.set_ylabel("Velocity (m/s)",color='purple')
fig.legend()

fig.tight_layout()
# solving the equation analytically yielded a result of 28.55686245854129 seconds, while my numerical result is 28.5568624585413 seconds. There is no significant difference in my two results.  
```

```python
g0 = 9.81  # (m/s^2) surface gravity
R = 6378.1*10**3  # (m) radius of Earth
y0 = 4000  # (m) depth of mineshaft

def Variablegravity(r):    # function for changing gravitational constant with distance
    return g0*(r/R)

def DerivativesVariableGravity(t, s):
    y, v = s
    dydt = v 
    dvdt = -Variablegravity(y) 
    return [dydt, dvdt]


def bottomofshaftVariableGravity(t, s):
    return s[0] - (R - y0)

bottomofshaftVariableGravity.terminal = False
bottomofshaftVariableGravity.direction = -1

v0 = 0   # intial velocity 
s0=[R,v0]

t_eval = np.linspace(0,30,500)
soln = solve_ivp(fun=DerivativesVariableGravity, y0=s0, t_span=(0,30), t_eval=t_eval, events=bottomofshaftVariableGravity)

tVariable_Gravity = soln.t
y1Variable_Gravity = soln.y[0]-(R-y0)
y2Variable_Gravity = soln.y[1]

t_event = soln.t_events[0][0]
print(t_event)

fig, ax = plt.subplots()
ax.plot(tVariable_Gravity,y1Variable_Gravity,color = 'blue',label="Position")
ax.set_xlabel("Time (s)")
ax2 = ax.twinx()
ax2.plot(tVariable_Gravity,y2Variable_Gravity,color = 'orange',label="Velocity")
ax.axhline(0,color = 'black')
ax.set_ylabel("Position (m)",color='blue')
ax2.set_ylabel("Velocity (m/s)",color='orange')
fig.legend()
ax.set_title("Free Fall with Variable Gravity (Figure 2)")
ax.grid()
fig.tight_layout()

# with the variable gravity I got a fall time to the bottom of that shaft 28.558355111904834 seconds, which is slighting longer than before, with normal ggravity
# at the small scale of the shaft however I would not expect this variable gravity too have that much of an effect


```

```python
#3
g0 = 9.81  # (m/s^2) surface gravity
R = 6378.1*10**3  # (m) radius of Earth
alpha = 0.0057#Drag coefficient
gamma = 2  
y0 = 4000  
def DerivativesVariableGravitywithDrag(t, s):
    y, v = s
    dydt = v
    dvdt = -Variablegravity(y) + alpha*(v)**gamma
    return [dydt, dvdt]
    
def bottomofshaftVariableGravitywithDrag(t, s):
    return s[0] - (R - y0)
    
bottomofshaftVariableGravitywithDrag.terminal = False
bottomofshaftVariableGravitywithDrag.direction = -1






v0 = 0   # intial velocity 
s0=[R,v0]

t_eval = np.linspace(0,100,500)
soln = solve_ivp(fun=DerivativesVariableGravitywithDrag, y0=s0, t_span=(0,100), t_eval=t_eval, events=bottomofshaftVariableGravitywithDrag)

tVariable_GravitywithDrag = soln.t
y1Variable_GravitywithDrag = soln.y[0]-(R-y0)
y2Variable_GravitywithDrag = soln.y[1]



t_event = soln.t_events[0][0]
print(t_event)


fig, ax = plt.subplots()
ax.plot(tVariable_GravitywithDrag,y1Variable_GravitywithDrag,color = 'blue',label="Position")
ax.set_xlabel("Time (s)")
ax2 = ax.twinx()
ax2.plot(tVariable_GravitywithDrag,y2Variable_GravitywithDrag,color = 'orange',label="Velocity")
ax.axhline(0,color = 'black')
ax.set_ylabel("Position (m)",color='blue')
ax2.set_ylabel("Velocity (m/s)",color='orange')
fig.legend()
ax.set_title("Free Fall with Variable Gravity with Drag")
ax.grid()
fig.tight_layout()
```

```python
# Part 3
CoriolisEffect = 7.272*10**-5  # rad/s
m = 1 #kg 
def CoriolisForce(t,s):
    x,y,vx,vy = s
    dxdt = vx
    dydt = vy
    dvxdt = 2*CoriolisEffect*vy
    dvydt = -g0-2*CoriolisEffect*vx
    return [dxdt, dydt, dvxdt, dvydt]


x0 = 0  # inital x position horizontally
y0 = 4000  # mineshaft height
vx0 = 0
vy0 = 0
s0 = [x0, y0, vx0, vy0]

def bottomofshaftCoriolis(t, s):
    return s[1]     # had to make sure to change this from s[0] to s[1] since now i have y as my second element in state vector
bottomofshaftCoriolis.terminal = True  # Changing to true this time for more clarity
bottomofshaftCoriolis.direction = -1


t_eval = np.linspace(0, 40, 500)


soln = solve_ivp(CoriolisForce,y0=s0, t_span=(0, 40), t_eval=t_eval, events=bottomofshaftCoriolis)
t = soln.t

x1C = soln.y[0]
y1C = soln.y[1]

fig, ax = plt.subplots()
ax.plot(x1C, y1C,label="Path with no drag",color='green')
ax.plot(x1C[::10], y1C[::10], 'o',color = 'black')
ax.set_xlabel("Transverse position (m)")
ax.set_ylabel("Depth (m)")
ax.set_title("Trajectory with Coriolis Force no drag  (Figure 3)")
ax.legend()
ax.grid()
fig.tight_layout()
print("Does it hit the wall?", max(abs(x1C)))

# We can see that the test object would hit the sides of the mineshaft, because the maximum is 5.526237811626339 meters
# and since the mineshaft is 5 meters wide and we drop from the center,we hit 2.5m much earlier than when we hit the bottom of the mineshaft



# with drag

alpha = 0.0057  # drag coefficient from earlier
def CoriolisWithDrag(t, s):
    x,y,vx,vy = s
    dxdt = vx
    dydt = vy
    dvxdt = 2*CoriolisEffect*vy-alpha*vx*abs(vx)
    dvydt = -g0-2*CoriolisEffect*vx-alpha*vy*abs(vy)
    return [dxdt, dydt, dvxdt, dvydt]

x0 = 0  # inital x position horizontally
y0 = 4000  # mineshaft height
vx0 = 0
vy0 = 0
s0 = [x0, y0, vx0, vy0]
t_eval = np.linspace(0, 100, 500)



solnDRAG = solve_ivp(CoriolisWithDrag,y0=s0, t_span=(0, 100), t_eval=t_eval, events=bottomofshaftCoriolis)


t = solnDRAG.t
xdrag = solnDRAG.y[0]
ydrag = solnDRAG.y[1]

fig, ax = plt.subplots()
ax.plot(xdrag, ydrag, label="Path with Drag",color = 'blue')
ax.plot(xdrag[::10], ydrag[::10], 'o',color = 'black')
ax.set_xlabel("Transverse position(m)")
ax.set_ylabel("Depth (m)")
ax.set_title("Trajectory with Coriolis and Drag")
ax.legend()
ax.grid()
fig.tight_layout()

print("Does it hit the wall with drag?", max(abs(xdrag)))
# with drag accounted for with the Corriolis affect 26.60702533594692 there is actually a significant difference in the tranverse position
# This can be accounted for by the drag causing the object to fall slower causing the corriolis force to a have a much greater affect on the trajectory of the object in free fall
```

```python
# Part 4

def InfinteMine(t, s):
    y, v = s
    dydt = v
    dvdt =  -Variablegravity(y) 
    return [dydt, dvdt]



s0 = [R, 0]
t_eval = np.linspace(0, 5000, 1000)


def centerofearth(t,s):
    return s[0]

soln = solve_ivp(InfinteMine, t_span=(0, 5000), y0=s0, t_eval=t_eval,events=centerofearth)
t = soln.t
y = soln.y[0]
v = soln.y[1]

centerofearthtime = soln.t_events[0][0]
print(centerofearthtime)
centerofearthspeed = soln.y_events[0][0][1]
print(centerofearthspeed)


fig, ax = plt.subplots()
ax.plot(t, y, label='Position (m)', color='blue')
ax.set_xlabel("Time (s)")
ax.set_ylabel("Position(m)", color='blue')
ax2 = ax.twinx()
ax2.plot(t, v, label='Velocity (m/s)', color='orange')
ax2.set_ylabel("Velocity (m/s)", color='orange')
ax.set_title("Infitely Deep Mine on Earth (Figure 4)")
ax.grid()
ax.legend()
fig.tight_layout()

# this graph is very interesting as it resembles harmonic motion like a spring or even an RLC circuit, this is caused by
# the object falling then reaching the center, then slowing down as it reaches the other side of the earth, then falling down again,
# then the process just keeps repeating. 

# it takes 2500 seconds for the object to reach the other side of the earth, and to reach the center of the earth it takes 1266.473469558803 seconds
# while the velocity is 7910.759872777071 downwards

# 3
G = 6.67430*10**-11      # m^3kg^-1s^-2  Gravitational constant
M = 5.972*10**24         # kg Mass of Earth


orbital = np.sqrt(G*M/R)

T_orbital = 2*np.pi*R/orbital     # orbital period

print(orbital)
print(T_orbital)

# the orbital period is  5069.371199675785 seconds which is about double the crossing time I got of 2500 seconds.  this means that the time it would take to fall through the entire earth`
# is porportional to the time it would take for 1 orbit. 
```

```python
# Part 5
from scipy.integrate import quad
r = np.linspace(0, R, 500) 
n_values = [0, 1, 2, 9]  
centerofearthspeedandtime = []   # for part 4

fig, ax = plt.subplots()

for n in n_values:
    rho_vals = (1 -(r/R)**2)**n
    ax.plot(r/R, rho_vals,label=f" n = {n}")

ax.set_xlabel("normalized Radius")
ax.set_ylabel("Density")
ax.set_title("Densities for different exponents  (Figure 5)")
ax.legend()
ax.grid()
fig.tight_layout()


# 2

M_earth  = 5.972*10**24 # kg Mass of the Earth
fig2, ax2 = plt.subplots()
for n in n_values:
    
    def rho2(r):
        return (1 -(r/R)**2)**n
    mass_integral, err = quad(lambda r: rho2(r)*r**2, 0, R)
    rhon2 = M_earth/(4*np.pi*mass_integral)


    F = []     # Calculating the force as a funciton of radius
    for rval in r:
        mr, err = quad(lambda rp: rhon2*rho2(rp)*rp**2, 0, rval)
        force = -(G*mr)/rval**2
        F.append(force)
    ax2.plot((r/R), F, label=f" n = {n}")
ax2.set_xlabel("Radius ")
ax2.set_ylabel("Gravitational Force (N/kg)")
ax2.set_title("Gravitational Force Profile (Figure 6)")
ax2.grid()
ax2.legend()
fig2.tight_layout()



#3
fig3, ax3 = plt.subplots()
ax4 = ax3.twinx()
s0 = [R, 0]
t_eval = np.linspace(0, 10000, 1000)


def centerofearth(t, s):
        return s[0] 
centerofearth.terminal = False
centerofearth.direction = -1

for n in n_values:    # defining my force function again same as above, then imputing it into the velocity and position functions. 
    def rho2(r):
        return (1 -(r/R)**2)**n   
        
    mass_integral, err = quad(lambda r: rho2(r)*r**2, 0, R)
    rhon2 = M_earth/(4*np.pi*mass_integral)
    
    def DerivativesnoDragDensity(t,s):
            y,v = s
            dydt = v
            mr, err = quad(lambda rp: rhon2 * rho2(rp) * rp**2, 0, y)
            force = G*mr/y**2
            dvdt = -force
            return [dydt,dvdt]
    def centerofearth(t, s):
        return s[0] 
    centerofearth.terminal = False
    centerofearth.direction = -1   

    solnDensity = solve_ivp(fun=DerivativesnoDragDensity,t_span=(0,10000),y0=s0,t_eval=t_eval,events=centerofearth)
    t=solnDensity.t
    y1 = solnDensity.y[0]
    y2 = solnDensity.y[1]
    ax3.plot(t,y1,label=f" Position for n ={n}")
    ax4.plot(t,y2,label=F" Velocity for n = {n}") 
    
    centertime = solnDensity.t_events[0][0]
    centervelocity = solnDensity.y_events[0][0][1]
    
    centerofearthspeedandtime.append((n, centertime, centervelocity))
for n, t, v in centerofearthspeedandtime:
    print(f"{n},{t}(s),{v} (m/s)")

ax3.legend(loc="upper right")
ax4.legend(loc="lower right")
ax3.set_xlabel("Time (s)")
ax3.set_ylabel("Position (m)",color = 'red')
ax4.set_ylabel("Velocity (m/s)",color='blue')
ax3.set_title("Position and Velocity with Realistic Density (Figure 7)")
fig3.tight_layout()





```
