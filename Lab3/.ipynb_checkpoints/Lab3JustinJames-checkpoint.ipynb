```python
# problem 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#constant


mz0 = 91.1880  # mass of Z boson GeV/c^2
mz0err = 0.0020 # error
mW = 80.3692 # mass of W boson GeV/c^2
mWerr = 0.0133
mH = 125.20 # mass of Higgs Boson GeV/c^2
mHerr = 0.11 
me = 0.51099895000 # mass of electron 
meerr = 0.00000000015
mmu = 105.6583755 # mass of muon MeV/c^2
mmuerr = 0.0000023
mr = 1776.93  # mass of tau MeV/c^2
mrerr = 0.09
data = np.loadtxt("atlas_z_to_ll.csv",delimiter=',',skiprows=1,)
pt1 = data[:,0]
pt2 = data[:,1]
eta1 = data[:,2]
eta2 = data[:,3]
phi1 = data[:,4]
phi2 = data[:,5]
E1 = data[:,6]
E2 = data[:,7]
PT = pt1 +pt2

px1 = pt1*np.cos(phi1)
px2 = pt2*np.cos(phi2)
py1 = pt1*np.sin(phi1)
py2 = pt2*np.sin(phi2)
pz1 = pt1*np.sinh(eta1)
pz2 = pt2*np.sinh(eta2)

px = px1+px2
py= py1+py2
pz = pz1+pz2
E = E1+E2


Mass = np.sqrt(E**2-(px**2+py**2+pz**2))
print(Mass)

bins = np.linspace(80,100,41)
fig, ax = plt.subplots(1,1)
counts, edges, =np.histogram(Mass,bins=bins)
ax.hist(Mass,bins=bins,color='skyblue', edgecolor='blue', alpha=0.6, label='Data')
centers = 0.5*(edges[:-1] + edges[1:])
yerr=np.sqrt(counts)
ax.errorbar(centers,counts,yerr=yerr,fmt='+',color='k')
ax.set_title("Invariant Mass")
ax.set_xlabel("Invariant Mass (GeV)")
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_ylabel("Events per Bin")
fig.tight_layout()
```

```python
# problem 2
from scipy.stats import chi2
mask = (centers > 87) & (centers < 93)
fit_x = centers[mask] 
fit_y = counts[mask]
fit_err = yerr[mask]
def D(m,mz0,gamma):
    A = 5000/2
    return (A/np.pi)*((gamma/2)/((m-mz0)**2+(gamma/2)**2))
params, pcov  = curve_fit(D,fit_x,fit_y, sigma=fit_err,absolute_sigma=True)
m_fit, Gamma_fit = params
m_err, Gamma_err = np.sqrt(np.diag(pcov))
m = np.linspace(80,100,40)

residuals = fit_y - D(fit_x, *params)
print(params[1])
fig, (ax_top, ax_bot) = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [3, 1]})
ax_top.errorbar(centers,counts,yerr=yerr,fmt='+',color='k')
ax_top.plot(m,D(m,*params),label= "Breit-Wigner Fit")
ax_top.hist(Mass,bins=bins)
ax_top.set_ylabel("Events per Bin")
ax_top.set_title("Figure 1: Invariant Mass and Breit-Wigner Fit")
ax_top.axvline(87, color='gray', linestyle='dotted')
ax_top.axvline(93, color='gray', linestyle='dotted')
ax_top.legend()
ax_top.set_xlabel("Rest Mass (GeV)")
ax_bot.errorbar(fit_x, residuals, yerr=fit_err, fmt='o', color='C1')
ax_bot.axhline(0)
ax_bot.set_xlabel("Invariant Mass (GeV)")
ax_bot.set_ylabel("Residuals")
fig.tight_layout()

chi2_val = np.sum((residuals / fit_err) ** 2)
ndof = len(fit_y) - len(params)
pval = chi2.sf(chi2_val, ndof)




print(f"Best-fit m_Z = {m_fit:.1f} +- {m_err:.1f} GeV")
print(f"Gamma = {Gamma_fit:.1f} +- {Gamma_err:.1f} GeV")
print(f"Chi-squared = {chi2_val:.1f}")
print(f"NDOF = {ndof}")
print(f"p-value = {pval:.1f}")
fig.savefig("figure_1.png", dpi=300)

```

```python
# Problem 3
n_bins = 300
mz_vals = np.linspace(89, 91, n_bins)
gamma_vals = np.linspace(5, 8, n_bins)

chi_map = np.zeros((n_bins, n_bins))
for i in range(n_bins):
    for j in range(n_bins):
        theory = D(fit_x, mz_vals[i], gamma_vals[j])
        chi2 = np.sum((fit_y - theory)**2 / fit_err**2)
        chi_map[j, i] = chi2 

chi_min = np.min(chi_map)
delta_chi_map = np.clip(chi_map, chi_min, chi_min + 35) - chi_min

X, Y = np.meshgrid(mz_vals, gamma_vals)

fig, ax = plt.subplots(figsize=(8, 6))
cs = ax.contourf(X, Y, delta_chi_map, 500, cmap="Blues_r")
cbar = fig.colorbar(cs, ax=ax)
cbar.set_label(r"$\Delta \chi^2$")
levels = [2.30, 11.83]
contours = ax.contour(X, Y, delta_chi_map, levels=levels,
                      colors=["yellow", "orange"],
                      linestyles=["solid", "dashed"])
ax.clabel(contours, fmt={2.30: "1 sigma", 11.83: "3 sigma"}, inline=True)

# Mark best fit point
ax.plot(m_fit, Gamma_fit, 'r+', markersize=10, label="Best Fit")

# Labels and save
ax.set_xlabel(r"$m_Z$ (GeV)")
ax.set_ylabel(r"$\Gamma$ (GeV)")
ax.set_title("Figure 2: chi-squared Contour Map")
ax.legend()
fig.tight_layout()
fig.savefig("figure_2.png", dpi=300)
```
